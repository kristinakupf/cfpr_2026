```python
import os
import shutil

# Define the base directory
base_dir = "/h/kupfersk/cfpr_2025/generate_forecasts/output/Full_Report/"
new_base_dir = base_dir.replace("Full_Report", "Forecasts")

# Iterate through all directories and subdirectories
for root, dirs, files in os.walk(base_dir):
    if "2024-07-01" in root:
        # Construct the new path by replacing 'Full_Report' with 'Forecasts'
        new_root = root.replace("Full_Report", "Forecasts")
        
        # Ensure the target directory exists
        os.makedirs(new_root, exist_ok=True)

        # Copy each file to the new path
        for file in files:
            src_file_path = os.path.join(root, file)
            dst_file_path = os.path.join(new_root, file)
            shutil.copy(src_file_path, dst_file_path)
            print(f"Copied {src_file_path} to {dst_file_path}")
```
