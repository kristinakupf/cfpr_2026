```python
import json
# Save the dictionary as a JSON file
with open('/h/kupfersk/cfpr_2025/generate_forecasts/experiment_dict.json', 'w') as json_file:
    json.dump(category_dict, json_file, indent=4)
```

```python
consolidated_dict = {}

for category, experiments in category_dict.items():
    # Add the "food cpi: " prefix to the category name
    category_with_prefix = f"food_cpi: {category}"
    
    for experiment in experiments:
        # Extract the experiment prefix (e.g., "exp7", "exp1_corr")
        experiment_parts = experiment.split('_')
        if len(experiment_parts) > 2:
            # Handles cases like "exp1_corr" (prefix includes two parts)
            experiment_prefix = '_'.join(experiment_parts[:2]) + '_'
        else:
            # Handles cases like "exp7" (prefix includes one part)
            experiment_prefix = experiment_parts[0] + '_'
        
        # Consolidate experiment prefixes in the new dictionary
        if experiment_prefix not in consolidated_dict:
            consolidated_dict[experiment_prefix] = []
        
        # Add the category with the prefix to the corresponding experiment prefix
        if category_with_prefix not in consolidated_dict[experiment_prefix]:
            consolidated_dict[experiment_prefix].append(category_with_prefix)
            
consolidated_dict

with open('/h/kupfersk/cfpr_2025/generate_forecasts/target_category_dict.json', 'w') as json_file:
    json.dump(consolidated_dict, json_file, indent=4)
```
