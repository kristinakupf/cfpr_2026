```python
import pandas as pd
import numpy as np
import matplotlib.colors as mc # For the legend
import matplotlib.pyplot as plt
import seaborn as sns
import math
from sklearn.preprocessing import MinMaxScaler

from data_functions import plot_data 
```

Livestock production index from World Bank

```python
df = pd.DataFrame(pd.read_csv("../cfpr_datasets/World_Bank_Livestock_production_index.csv", encoding= 'unicode_escape'))
df = df.iloc[[3,39]]
df = df.dropna(axis=1)
df = df.iloc[:, 4:]
df = df.transpose()
df.head(10)

df = df.rename(columns= {df.columns[0]: 'Date', df.columns[1]: "Livestock Index"})
df = df.set_index('Date')
df

plot_data(df, 'Livestock Index')


file_name = 'LivestockIndex.csv'
df.to_csv(file_name, index=True)
```
