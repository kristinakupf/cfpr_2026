US Natural Gas Export Prices in 1000 USD from US Energy Information Administration

```python
### US Natural Gas Export Prices in 1000 USD
 #### from US Energy Information Administration

df = pd.DataFrame(pd.read_csv("../cfpr_datasets/NG_PRI_SUM_DCU_NUS_M.csv", encoding= 'unicode_escape'))
df = df.drop(df.index[0:2])
# df5.info()
df = df[['Back to Contents', 'Unnamed: 4']]
df = df.dropna()
df = df.rename(columns= {"Back to Contents": 'Date', "Unnamed: 4": "US Natural Gas Price Exports"})

# changes the object type to numeric
df['US Natural Gas Price Exports']=pd.to_numeric(df['US Natural Gas Price Exports'])

## Setting date as index
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index('Date')
df

plot_data(df['US Natural Gas Price Exports'], 'Prices of US Natural Gas Exports [Dollar/Thousand Cubic feet')


file_name = 'Natural_Gas.csv'
df.to_csv(file_name, index=True)
```
