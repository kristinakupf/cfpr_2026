```python
## Google Trends

#food safety
df7 = pd.DataFrame(pd.read_csv("/home/varsallz/cfprS24_z/cfpr_data_collection/data/raw_data/GT_food_safety.csv", encoding= 'unicode_escape')).dropna()
df7 = df7.rename(columns={"Category: All categories": "Date", "Unnamed: 1" : "unsafe food", "Unnamed: 2": "food quality control"})
df7 = df7.drop(df7.index[0])
df7['Date'] = pd.to_datetime(df7['Date'], format= "%b-%y")
df7 = df7.set_index('Date')
df7.replace('<1', 0.0001, inplace=True)

#food security
df8 = pd.DataFrame(pd.read_csv(raw_path+"GT_food_security.csv", encoding= 'unicode_escape')).dropna()
df8 = df8.rename(columns={"Category: All categories": "Date", "Unnamed: 1" : "food bank", "Unnamed: 2": "free food"})
df8 = df8.drop(df8.index[0])
df8['Date'] = pd.to_datetime(df8['Date'], format= "%b-%y")
df8 = df8.set_index('Date')
df8.replace('<1', 0.0001, inplace=True)

#dieting
df9 = pd.DataFrame(pd.read_csv(raw_path+"GT_dieting.csv", encoding= 'unicode_escape')).dropna()
df9 = df9.drop(df9.index[0])
df9 = df9.rename(columns={"Category: All categories": "Date", "Unnamed: 1" : "diet", "Unnamed: 2": "junk food",  "Unnamed: 3": "organic food",  "Unnamed: 4": "healthy diet"})

df9['Date'] = pd.to_datetime(df9['Date'], format= "%b-%y")
df9 = df9.set_index('Date')
df9.replace('<1', 0.0001, inplace=True)

# #avian flu
df10 = pd.DataFrame(pd.read_csv(raw_path+"GT_avian_flu.csv", encoding= 'unicode_escape')).dropna()
df10 = df10.rename(columns={"Category: All categories": "Date", "Unnamed: 1" : "avian_flu", "Unnamed: 2": "bird flu",  "Unnamed: 3": "flu",  "Unnamed: 4": "zoonotic disease"})
df10 = df10.drop(df10.index[0])
df10['Date'] = pd.to_datetime(df10['Date'], format= "%b-%y")
df10 = df10.set_index('Date')
df10.replace('<1', 0.0001, inplace=True)


scaler = MinMaxScaler(feature_range=(0, 1))
df7 = pd.DataFrame(scaler.fit_transform(df7), columns=df7.columns, index=df7.index)
df8 = pd.DataFrame(scaler.fit_transform(df8), columns=df8.columns, index=df8.index)
df9 = pd.DataFrame(scaler.fit_transform(df9), columns=df9.columns, index=df9.index)
df10 = pd.DataFrame(scaler.fit_transform(df10), columns=df10.columns, index=df10.index)

```

```python
## Cross Correlation
# Concatenate the DataFrames along the columns axis
concatenated_df = pd.concat([df7,df8, df9, df10], axis=1)
start = '2018-04-01'
end = '2024-04-01'
concatenated_df= concatenated_df.loc[start:end]

# # exogeneous and lagged data
combined_df = pd.concat([df_lagged, concatenated_df], axis=1).dropna()
combined_df
corr_matrix = combined_df.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.title('Cross-correlation Heatmap')
plt.show()
```
