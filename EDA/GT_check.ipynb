```python
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
file_path = '../data/raw_data/GT_food_security.csv'
combined_df = pd.read_csv(file_path, index_col=0)
combined_df = combined_df.iloc[1:, :]
combined_df = combined_df.apply(pd.to_numeric, errors='coerce')


scaler = StandardScaler()

# Scale the DataFrame
scaled_df_combined = pd.DataFrame(scaler.fit_transform(combined_df), columns=combined_df.columns, index=combined_df.index)
scaled_df_combined
```

```python
directory = '../data/check_GT_data/'

# Dictionary to hold DataFrames
# dfs_dict = {}
dfs = []

# Read and store each CSV file in the dictionary
for filename in os.listdir(directory):
    file_path = os.path.join(directory, filename)
    df = pd.read_csv(file_path, index_col=0)
    df = df.iloc[1:, :]
    dfs.append(df)

concatenated_df  = pd.concat(dfs, axis=1,ignore_index=False)
concatenated_df = concatenated_df.apply(pd.to_numeric, errors='coerce')

scaler = StandardScaler()

# Scale the DataFrame
scaled_df = pd.DataFrame(scaler.fit_transform(concatenated_df), columns=concatenated_df.columns, index=concatenated_df.index)




scaled_df
# df
```
